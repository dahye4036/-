구매 금액의 z-score 계산하기

```python
# z-score 계산을 위한 실습용 데이터

temp = pd.DataFrame({'a': [10,20,30,20,10,200], 'b': [100,300,200,150,250,200], 'c': [10, 500, 20, 250, 25, 200]})
```

```python
# temp 데이터의 컬럼별 표준편차를 계산

temp.std()
```

```python
# 각각의 데이터에 대해 a 컬럼의 z-score 계산
# (data - mean) / std

(temp['a'] - 48.33) / 74.67
```

```python
# 각각의 데이터에 대해 b 컬럼의 z-score 계산

(temp['b'] - 48.33) / 74.67
```

```python
# 각각의 데이터에 대해 c 컬럼의 z-score 계산

(temp['c'] - 48.33) / 74.67
```

```python
# cc_num 컬럼의 값마다 데이터 개수 계산

cc_df['cc_num'].value_counts()
```

```python
# cc_num 컬럼의 값마다 amt의 평균, 표준편차를 계산하여 amt_info에 저장

amt_info = cc_df.groupby('cc_num')['amt'].agg(['mean', 'std']).reset_index()
```

```python
amt_info.to_pickle('./amt_info.pkl')
```

```python
# cc_num 컬럼을 기준으로, cc_df와 amt_info 데이터를 합쳐서 cc_df에 저장
# (left merge 수행)

cc_df = pd.merge(cc_df, amt_info, how='left', on='cc_num')
```

```python
# 결제금액(amt)의 z-score를 계산하고, amt_z 컬럼에 저장
# 평균: mean, 표준편차: std

amt_z = (cc_df['amt'] - cc_df['amt'].mean()) / cc_df['amt'].std()
```

```python
# 사기 거래 데이터(is_fraud=1)를 확인

cc_df[cc_df['is_fraud']==1]
```

```python
# z-score 계산 완료. 이제 mean, std 컬럼 제거. 

cc_df.drop(['mean', 'std'], axis=1, inplace=True)
```

```python
# cc_num, category 컬럼의 값마다 결제금액(amt)의 평균, 표준편차 계산. cat_info에 저장. 

cat_info = cc_df.groupby(['cc_num', 'category'])['amt'].agg(['mean', 'std']).reset_index()
```

```python
cat_info.to_pickle('./cat_info.pkl')
```

```python
# cc_num 컬럼을 기준으로, cc_df와 cat_info 데이터를 합쳐서 cc_df에 저장
# (left merge 수행)

cc_df = pd.merge(cc_df, cat_info, how='left', on='cc_num')
```

```python
# 결제금액(amt)의 z-score을 계산하여 cat_amt_z 컬럼에 저장
# 평균: mean, 표준편차: std

cat_amt_z = (cc_df['amt'] - cc_df['amt'].mean()) / cc_df['amt'].std()
```

결제 시간 관련 feature 분석

```python
# trans_date_trans_time 컬럼에서 시간 값 추출하여 hour 컬럼에 저장
# (힌트: datetime 자료형 이용)

cc_df['hour'] = pd.to_datetime(cc_df['trans_date_trans_time']).dt.hour
```

```python
cc_df.head()
```

```python
# 결제시간을 mornig, afternoon, night, evening으로 분류하기 위해 함수 정의

def hour_func(x):
    if (x>=6) & (x<18):
        return 'morning'
    elif (x>=12) & (x<18): 
        return 'afternoon'
    elif (x>=18) & (x<23):
        return 'night'
    else:
        return 'evening'
```

```python
# hour 컬럼에 hour_func 함수를 적용한 값을 hour_cat 컬럼에 저장

hour_cat = cc_df['hour'].apply(hour_func)
```

```python
cc_df.head()
```

```python
# 시간대별로 데이터 수를 계산하여 출력

cc_df.groupby('hour')['cc_num'].value_counts()
```

```python
# cc_num 컬럼의 값마다 amt 컬럼의 데이터 수를 계산하여 all_cnt 변수로 저장

all_cnt = cc_df.groupby('cc_num')['amt'].count().reset_index()
```

```python
# cc_num, hour_cat 컬럼의 값마다 amt 컬럼의 데이터 수를 계산하여 hour_cnt 변수로 저장

hour_cnt = cc_df.groupby(['cc_num', 'hour_cat'])['amt'].count().reset_index()

# 이게 왜 에러인데 ㅠ 
```

```python
all_cnt.head()
```

```python
hour_cnt.head()

# 위에서 hour_cnt 도출 안되니까 밑에도 주르르르르륵... 에러 뜸!! 
```

```python
# cc_num 컬럼 기준으로 hour_cnt와 all_cnt 합치기
# (hour_cnt에서 left merge 수행)

hour_cnt = pd.merge(hour_cnt, all_cnt, how='left', on='cc_num')
```

```python
hour_cnt.head()
```

```python
# amt_x와 amt_y 컬럼의 이름을 각각 hour_cnt, total_cnt로 변경

cc_df = cc_df.rename({'amt_x': 'hour_cnt', 'amt_y': 'total_cnt'})
```

```python
hour_cnt.head()
```

```python
# 전체 거래 건수(total_cnt)에 대한 시간대별 거래 건수(hour_cnt)의 비율 계산
# hour_perc 컬럼에 저장

cc_df['hour_perc'] = cc_df.groupby('hour')['hour_cnt'].value_counts() / cc_df.groupby('hour')['total_cnt'].value_counts
```
